```python
import pandas as pd
a = pd.read_csv('training_set_features.csv')
```

```python
a.head()
```

## Removing NaN

```python
a = a.apply(lambda x: x.fillna(x.mode()[0]))
```

```python
a.isnull().sum()
```

## OneHot Encoding

```python
a = pd.get_dummies(a, columns=['race', 'sex', 'marital_status', 'rent_or_own', 'employment_status', 'hhs_geo_region', 'census_msa', 'employment_industry', 'employment_occupation'])
```
